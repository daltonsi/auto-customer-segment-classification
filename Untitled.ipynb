# Customer Segment Classification

## Getting the data

#### Data Summary

```python
import pandas as pd
pd.set_option("display.max_rows", 100)


train = pd.read_csv('raw_data/Train.csv')
y_train, X_train = train.loc[:, train.columns == 'Segmentation'].copy(), train.drop('Segmentation', axis=1)
X_Train_exp = X_train.copy() #

X_train.drop('ID',axis=1,inplace=True)

test = pd.read_csv('raw_data/Test.csv')
y_test, X_test = test.loc[:, test.columns == 'Segmentation'].copy(), test.drop('Segmentation', axis=1)

X_train.head()
```

#### Null Handling

```python
# Missing values per feature
X_train.isnull().sum(axis = 0)
```

```python
X_train.isnull().sum(axis=1).value_counts()
```

```python
# Remove NA rows for features with low NA row counts (< 1% of total raw row count)
X_train.dropna(how='any', subset=['Graduated'], axis=0, inplace=True)
X_train.dropna(how='any', subset=['Var_1'], axis=0, inplace=True)

# Impute Boolean Features with Mode
X_train['Ever_Married'].fillna(X_train['Ever_Married'].mode()[0], inplace=True)


# Impute WE with median at ~1 as mean (2.64) is skewed towards 75th percentile (4)
X_train['Work_Experience'].fillna(X_train['Work_Experience'].median(), inplace=True)


# Impute Family size with mean (2.85) which aligns close to median (3)
X_train['Family_Size'].fillna(X_train['Family_Size'].mean(), inplace=True)


# NA values of Profession will report as all zeros post one hot encoding - no need to prefilter
```

```python
X_train['Profession'].value_counts(dropna=False)
```

```python
X_train.isnull().sum(axis = 0)
```

#### Preprocessing boolean  variables

```python
X_train['Ever_Married'] = X_train['Ever_Married'].map({'No': False, 'Yes': True})
X_train['Graduated'] = X_train['Graduated'].map({'No': False, 'Yes': True})

X_train.head()
```

#### One Hot encode categorical variables

```python
from sklearn.preprocessing import OneHotEncoder


    
# Gender
gender_enc = OneHotEncoder()
gen_encoded = gender_enc.fit_transform(X_train[['Gender']]) # returns df of included col indcs
X_train[gender_enc.categories_[0]] = gen_encoded.toarray()

# Profession
prof_enc = OneHotEncoder(handle_unknown='ignore')
prof_encoded = prof_enc.fit_transform(X_train[['Profession']])
X_train[['prof_'+str(i) for i in prof_enc.categories_[0]]] = prof_encoded.toarray()

# Spending Score
ss_enc = OneHotEncoder()
ss_encoded = ss_enc.fit_transform(X_train[['Spending_Score']]) 
X_train[['ss_'+ i for i in ss_enc.categories_[0]]] = ss_encoded.toarray()

var1_enc = OneHotEncoder()
var1_encoded = var1_enc.fit_transform(X_train[['Var_1']]) 
X_train[var1_enc.categories_[0]] = var1_encoded.toarray()



X_train.drop(['prof_nan','Profession','Gender','Spending_Score','Var_1'],axis=1, inplace=True)
    
    
    

```

#### Scale the continous features




```python
from sklearn.preprocessing import StandardScaler

#Age Scaler
age_scaler = StandardScaler()
age_scaled = age_scaler.fit_transform(X_train[['Age']])
X_train['Age'] = age_scaled



#Work Experience
work_scaler = StandardScaler()
work_scaled = work_scaler.fit_transform(X_train[['Work_Experience']])
X_train['Work_Experience'] = work_scaled


#Family Size
fam_scaler = StandardScaler()
fam_scaled = fam_scaler.fit_transform(X_train[['Family_Size']])
X_train['Family_Size'] = fam_scaled

X_train.head()


```
